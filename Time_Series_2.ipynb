### Setting the Desired Frequency

```python
df_comp=df_comp.asfreq('d')
#h- hourly
#d - daily
#b - business days
#w - weekly
#m - monthly
#a - yearly(annual)
```

```python
df_comp=df_comp.asfreq('b')
```

### Handling Missing Values

```python
df_comp.isna()
```

```python
df_comp.isna().sum()
#setting freq to business days must have added 8 additional days
```

```python
df_comp
```

```python
df_comp.spx=df_comp.spx.fillna(method='ffill')
```

```python
#df_comp.isna().sum()
```

```python
df_comp.ftse=df_comp.ftse.fillna(method='bfill')
```

```python
df_comp.dax=df_comp.dax.fillna(value = df_comp.dax.mean())
#filling missing values with mean in time series might
  #disturb the flow
```
